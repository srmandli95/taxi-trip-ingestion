```python
import pandas as pd
import json
from pathlib import Path

# data folder from fetch_all.py
DATA = Path("data/raw")
```

```python
trips_path = DATA / "trips/yellow/2024/yellow_tripdata_2024-01.parquet"
df_trips = pd.read_parquet(trips_path)
df_trips.info()
print("columns", df_trips.columns)
print("Trips shape:", df_trips.shape)
display(df_trips.head(25))
display(df_trips[df_trips.isnull().any(axis=1)])
print(df_trips.isnull().sum())
```

```python
zones_path = DATA / "geo/taxi_zone_lookup.csv"
df_zones = pd.read_csv(zones_path)
print("Zones shape:", df_zones.shape)
display(df_zones.head(200))
```

```python
weather_path = DATA / "weather/openmeteo_hourly_2024-01-01_2024-01-31.json"
with open(weather_path) as f:
    weather_raw = json.load(f)

df_weather = pd.DataFrame({
    "ts_utc": weather_raw["hourly"]["time"],
    "temp_c": weather_raw["hourly"]["temperature_2m"],
    "precip_mm": weather_raw["hourly"]["precipitation"],
})
df_weather["ts_utc"] = pd.to_datetime(df_weather["ts_utc"])
print("Weather shape:", df_weather.shape)
display(df_weather.head())
```

```python

pois_path = DATA / "pois/osm_pois_manhattan.json"

with open(pois_path) as f:
    pois_raw = json.load(f)

# Flatten JSON to a DataFrame
df_pois = pd.json_normalize(pois_raw["elements"])

print("POIs shape:", df_pois.shape)
display(df_pois.head(10))
```
